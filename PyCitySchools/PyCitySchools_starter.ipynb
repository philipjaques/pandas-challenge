```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
### Create a series of variables for cleaner calculation

# Total number of schools
total_number_schools = school_data_complete_df["school_name"].nunique()

# Total students
total_students = school_data_complete_df["Student ID"].nunique()

# Total budget
total_budget = school_data_complete_df["budget"].sum()

# Average math score
average_math_score = school_data_complete_df["math_score"].mean()

# Average reading score
average_reading_score = school_data_complete_df["reading_score"].mean()

# Passing Math
passing_math = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70]

# Percent passing math
percent_passing_math = (len(passing_math) / school_data_complete_df["math_score"].count()) * 100

# Passing Reading
passing_reading = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70]

# Percent passing reading
percent_passing_reading = (len(passing_reading) / school_data_complete_df["reading_score"].count()) * 100

# Percent passing math AND reading
percent_passing_math_AND_reading = (len(school_data_complete_df[(school_data_complete_df["math_score"]>=70) & (school_data_complete_df["reading_score"]>=70)]) / school_data_complete_df["reading_score"].count()) * 100

district_summary_df = pd.DataFrame({
    "Total Number of Schools": [total_number_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Percent Passing Math": [percent_passing_math],
    "Percent Passing Reading": [percent_passing_reading],
    "Percent Passing Math & Reading": [percent_passing_math_AND_reading]
})

# Better formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,.0f}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["Percent Passing Math"] = district_summary_df["Percent Passing Math"].map("{:.1f}%".format)
district_summary_df["Percent Passing Reading"] = district_summary_df["Percent Passing Reading"].map("{:.1f}%".format)
district_summary_df["Percent Passing Math & Reading"] = district_summary_df["Percent Passing Math & Reading"].map("{:.1f}%".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python

### Create a series of variables for cleaner calculation

# Index on school name and type
school_types = school_data.set_index(["school_name"])["type"]

#Total students per school
students_per_school = school_data_complete_df["school_name"].value_counts()

#Total budget per school
budget_per_school = school_data_complete_df.groupby(["school_name"])["budget"].mean()

#Budget per student
budget_per_student = budget_per_school/students_per_school

#Average math score per school
average_math_per_school = school_data_complete_df.groupby(["school_name"])["math_score"].mean()

#Average reading score per school
average_reading_per_school = school_data_complete_df.groupby(["school_name"])["reading_score"].mean()

# Get a new dataframe for only passing math scores
passing_math_per_school_df = school_data_complete_df.loc[school_data_complete_df["math_score"]>=70]

# Get the counts of passing math scores per school
count_of_passing_math_per_school = passing_math_per_school_df["school_name"].value_counts()

# Get the percentage passing math per school
percent_passing_math_per_school = (count_of_passing_math_per_school / students_per_school) * 100

# Get a new dataframe for only passing reading scores
passing_reading_per_school_df = school_data_complete_df.loc[school_data_complete_df["reading_score"]>=70]

# Get the counts of passing reading scores per school
count_of_passing_reading_per_school = passing_reading_per_school_df["school_name"].value_counts()

# Get the percentage passing reading per school
percent_passing_reading_per_school = (count_of_passing_reading_per_school / students_per_school) * 100

# Percentage of students passing math AND reading per school
percent_passing_math_AND_reading_per_school = ((count_of_passing_math_per_school + count_of_passing_reading_per_school) / students_per_school) * 100

# Create a dataframe to hold the school summary
school_summery_df = pd.DataFrame({  
    "School Type": school_types, 
    "Total Students": students_per_school, 
    "Total School Budget": budget_per_school,
    "Per Student Budget": budget_per_student,
    "Average Math Score": average_math_per_school,
    "Average Reading Score": average_reading_per_school,
    "% Passing Math": percent_passing_math_per_school, 
    "% Passing Reading": percent_passing_reading_per_school,
    "% Overall Passing": percent_passing_math_AND_reading_per_school
})
school_summery_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# sort descending
top_performaing_schools_df = school_summery_df.sort_values(["% Overall Passing"], ascending=False)
top_performaing_schools_df
```

## Math Scores by Grade

  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
### Create variables

# Create new dataframe for only 9th grade scores
ninth_grade_df = school_data_complete_df.loc[school_data_complete_df["grade"] == "9th"].groupby(["school_name"])

# Get the ninth grade math scores
ninth_grade_math = ninth_grade_df["math_score"].mean()

# Create new dataframe for only 10th grade scores
tenth_grade_df = school_data_complete_df.loc[school_data_complete_df["grade"] == "10th"].groupby(["school_name"])

# Get the tenth grade math scores
tenth_grade_math = tenth_grade_df["math_score"].mean()

# Create new dataframe for only 11th grade scores
eleventh_grade_df = school_data_complete_df.loc[school_data_complete_df["grade"] == "11th"].groupby(["school_name"])

# Get the eleventh grade math scores
eleventh_grade_math = eleventh_grade_df["math_score"].mean()

# Create new dataframe for only 12th grade scores
twelfth_grade_df = school_data_complete_df.loc[school_data_complete_df["grade"] == "12th"].groupby(["school_name"])

# Get the twelfth grade math scores
twelfth_grade_math = twelfth_grade_df["math_score"].mean()

# Create a dataframe to hold the math scores by grade and map formatting inline
math_scores_by_grade = pd.DataFrame({   
    "9th Grade": ninth_grade_math.map("{:.1f}".format),
    "10th Grade": tenth_grade_math.map("{:.1f}".format),
    "11th Grade": eleventh_grade_math.map("{:.1f}".format),
    "12th Grade": twelfth_grade_math.map("{:.1f}".format)
})

math_scores_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create a new dataframe with only the columns we want to consider
school_summary_by_spending_df = school_summery_df.drop(columns=["School Type","Total Students","Total School Budget","Per Student Budget"])
school_summary_by_spending_df['Average Budget per Student'] = budget_per_student
new_order=[5,4,0,1,2,3]
school_summary_by_spending_df = school_summary_by_spending_df[school_summary_by_spending_df.columns[new_order]]

# Create bins
spending_bins = [0, 585, 615, 645, 675]
spending_group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# Cut the data
school_summary_by_spending_df["Spending Ranges"] = pd.cut(school_summary_by_spending_df["Average Budget per Student"], spending_bins, labels=spending_group_names)
scores_by_spending_df = school_summary_by_spending_df.groupby("Spending Ranges")

# Format data
school_summary_by_spending_df["Average Budget per Student"] = school_summary_by_spending_df["Average Budget per Student"].map("${:,.2f}".format)
scores_by_spending_df.max()
```
